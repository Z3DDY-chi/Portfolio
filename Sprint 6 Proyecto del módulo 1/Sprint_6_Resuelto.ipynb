```python
# Proyecto Ice Games - Análisis de Videojuegos
# Paso 2: Preparar los datos

# ====================================
# 1. REEMPLAZAR NOMBRES DE COLUMNAS (MINÚSCULAS)
# ====================================

print("\n" + "="*70)
print("1. NORMALIZACIÓN DE NOMBRES DE COLUMNAS")
print("="*70)

print("\n Nombres originales:")
print(df.columns.tolist())

# Convertir a minúsculas
df.columns = df.columns.str.lower()

print("\n Nombres normalizados (minúsculas):")
print(df.columns.tolist())

# ====================================
# 2. ANÁLISIS DETALLADO DE TIPOS DE DATOS ACTUALES
# ====================================

print("\n" + "="*70)
print("2. ANÁLISIS DE TIPOS DE DATOS ACTUALES")
print("="*70)
print(df.dtypes)
print(f"\n Info del DataFrame:")
print(df.info())

# ====================================
# 3. CONVERSIÓN DE TIPOS DE DATOS
# ====================================

print("\n" + "="*70)
print("3. CONVERSIÓN DE TIPOS DE DATOS")
print("="*70)

# 3.1 Analizar user_score antes de convertir
print("\n Análisis de 'user_score' ANTES de conversión:")
print(f"   - Tipo actual: {df['user_score'].dtype}")
print(f"   - Valores únicos: {df['user_score'].nunique()}")
print(f"   - Muestra de valores:")
print(df['user_score'].value_counts().head(10))

# Contar valores 'tbd'
tbd_count = (df['user_score'] == 'tbd').sum()
print(f"\n  Valores 'tbd' encontrados: {tbd_count}")

# 3.2 Convertir user_score a numérico
# Primero reemplazar 'tbd' por NaN, luego convertir a float
print("\n Convirtiendo 'user_score' a tipo numérico...")
df['user_score'] = pd.to_numeric(df['user_score'], errors='coerce')

print(" Conversión completada")
print(f"   - Nuevo tipo: {df['user_score'].dtype}")
print(f"   - Valores nulos después de conversión: {df['user_score'].isnull().sum()}")

# 3.3 Convertir year_of_release a entero (manejando NaN)
print("\n Convirtiendo 'year_of_release' a tipo entero...")
# Mantener como float por los NaN, pero es un año
print(f"   - Tipo actual: {df['year_of_release'].dtype}")
print("     Se mantiene como float64 para preservar valores NaN")

# ====================================
# 4. RESUMEN DE CAMBIOS EN TIPOS DE DATOS
# ====================================

print("\n" + "="*70)
print(" RESUMEN DE CAMBIOS EN TIPOS DE DATOS")
print("="*70)
print("""
 user_score: object → float64
  Razón: Es una puntuación numérica (0-10). Los valores 'tbd' se convirtieron 
  a NaN para permitir cálculos estadísticos.

 year_of_release: float64 (sin cambios)
  Razón: Se mantiene como float64 para preservar los valores NaN. 
  Representa años, pero algunos juegos no tienen fecha de lanzamiento.

 Columnas de ventas (na_sales, eu_sales, jp_sales, other_sales): float64
  Razón: Ya están en el formato correcto (valores decimales en millones de USD).

 critic_score: float64
  Razón: Ya está en formato correcto (puntuaciones de 0-100).

 Columnas categóricas (name, platform, genre, rating): object
  Razón: Son variables de texto/categoría, formato correcto.
""")

# ====================================
# 5. TRATAMIENTO DE VALORES AUSENTES
# ====================================

print("\n" + "="*70)
print("4. ANÁLISIS Y TRATAMIENTO DE VALORES AUSENTES")
print("="*70)

# Análisis actualizado de valores nulos
print("\n Valores nulos por columna (DESPUÉS de conversiones):")
null_summary = pd.DataFrame({
    'Valores_Nulos': df.isnull().sum(),
    'Porcentaje': (df.isnull().sum() / len(df)) * 100
})
print(null_summary)

# Visualización actualizada
plt.figure(figsize=(12, 6))
null_percentage = (df.isnull().sum() / len(df)) * 100
null_percentage.sort_values(ascending=False).plot(kind='bar', color='coral')
plt.title('Porcentaje de Valores Nulos por Columna (Después de Limpieza)', 
          fontsize=14, fontweight='bold')
plt.xlabel('Columnas', fontsize=12)
plt.ylabel('Porcentaje (%)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.axhline(y=50, color='red', linestyle='--', alpha=0.5, label='50%')
plt.legend()
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("\n" + "="*70)
print(" ESTRATEGIA PARA VALORES AUSENTES")
print("="*70)

print("""
 name (2 valores nulos, 0.01%):
   Decisión: ELIMINAR filas
   Razón: Sin nombre no podemos identificar el juego. Son muy pocos casos.

 genre (2 valores nulos, 0.01%):
   Decisión: ELIMINAR filas
   Razón: El género es fundamental para el análisis. Son muy pocos casos.

 year_of_release (269 valores nulos, 1.61%):
   Decisión: ELIMINAR filas
   Razón: El año es crucial para análisis temporal y proyecciones 2017.
   Sin año no podemos determinar si el juego es relevante para nuestro modelo.

 critic_score (8,578 valores nulos, 51.32%):
   Decisión: MANTENER valores nulos
   Razón: Es común que juegos antiguos o de menor importancia no tengan 
   reseñas profesionales. Eliminarlos reduciría mucho el dataset.
   Los usaremos cuando estén disponibles para análisis de correlación.

 user_score (7,590 valores nulos después de conversión, 45.41%):
   Decisión: MANTENER valores nulos
   Razón: Similar a critic_score. Muchos juegos no tienen puntuaciones de 
   usuarios. Los usaremos para análisis específicos cuando estén disponibles.

 rating (6,766 valores nulos, 40.48%):
   Decisión: MANTENER valores nulos
   Razón: Las clasificaciones ESRB no siempre están disponibles, especialmente
   para juegos antiguos o de mercados sin esta clasificación. Analizaremos
   su impacto solo con los datos disponibles.

 ¿Por qué hay valores ausentes?
   - Juegos antiguos: No existía información digital completa
   - Juegos menos populares: No fueron reseñados por críticos/usuarios
   - Datos incompletos en fuentes originales
   - Diferencias regionales en clasificación (Rating)
   - Valores 'tbd' en user_score: Juegos sin suficientes puntuaciones
""")

# Aplicar tratamiento
print("\n Aplicando tratamiento de valores ausentes...")

# Contar filas antes
filas_antes = len(df)
print(f"   Filas antes de limpieza: {filas_antes:,}")

# Eliminar filas con valores nulos en columnas críticas
df_clean = df.dropna(subset=['name', 'genre', 'year_of_release']).copy()

filas_despues = len(df_clean)
filas_eliminadas = filas_antes - filas_despues

print(f"   Filas después de limpieza: {filas_despues:,}")
print(f"   Filas eliminadas: {filas_eliminadas} ({(filas_eliminadas/filas_antes)*100:.2f}%)")

# Verificar que no hay nulos en columnas críticas
print("\n Verificación de columnas críticas:")
print(f"   - name: {df_clean['name'].isnull().sum()} nulos")
print(f"   - genre: {df_clean['genre'].isnull().sum()} nulos")
print(f"   - year_of_release: {df_clean['year_of_release'].isnull().sum()} nulos")

# ====================================
# 6. CALCULAR VENTAS TOTALES
# ====================================

print("\n" + "="*70)
print("5. CÁLCULO DE VENTAS TOTALES")
print("="*70)

# Crear columna total_sales
df_clean['total_sales'] = (df_clean['na_sales'] + 
                            df_clean['eu_sales'] + 
                            df_clean['jp_sales'] + 
                            df_clean['other_sales'])

print(" Columna 'total_sales' creada exitosamente")
print(f"\n Estadísticas de ventas totales:")
print(df_clean['total_sales'].describe())

print(f"\n Top 10 juegos por ventas totales:")
top_games = df_clean.nlargest(10, 'total_sales')[['name', 'platform', 'year_of_release', 'total_sales']]
print(top_games.to_string(index=False))

# ====================================
# 7. VERIFICACIÓN FINAL
# ====================================

print("\n" + "="*70)
print("6. VERIFICACIÓN FINAL DEL DATASET LIMPIO")
print("="*70)

print(f"\n Dimensiones finales: {df_clean.shape}")
print(f"\n Columnas finales:")
print(df_clean.columns.tolist())

print(f"\n Tipos de datos finales:")
print(df_clean.dtypes)

print(f"\n Valores nulos restantes:")
null_final = pd.DataFrame({
    'Valores_Nulos': df_clean.isnull().sum(),
    'Porcentaje': (df_clean.isnull().sum() / len(df_clean)) * 100
})
print(null_final)

print("\n" + "="*70)
print(" RESUMEN DEL PASO 2")
print("="*70)
print(f"""
 Nombres de columnas normalizados a minúsculas
 user_score convertido de object a float64 (valores 'tbd' → NaN)
 Filas con valores nulos críticos eliminadas: {filas_eliminadas} ({(filas_eliminadas/filas_antes)*100:.2f}%)
 Dataset final: {df_clean.shape[0]:,} filas × {df_clean.shape[1]} columnas
 Columna 'total_sales' agregada exitosamente
 Valores nulos restantes solo en columnas no críticas (scores, rating)

 Dataset listo para análisis exploratorio (Paso 3)
""")

# Guardar el dataset limpio para próximos pasos
print("\n Guardando dataset limpio como 'df_clean' en memoria...")
print("   Usa 'df_clean' para los siguientes análisis")

# Mostrar primeras filas del dataset limpio
print("\n" + "="*70)
print("VISTA PREVIA DEL DATASET LIMPIO")
print("="*70)
print(df_clean.head(10))

print("\n" + "="*70)
print(" PASO 2 COMPLETADO")
print("="*70)
print("Siguiente paso: Análisis de los datos (Step 3)\n")
```

```python
# Proyecto Ice Games - Análisis de Videojuegos
# Paso 3: Analizar los datos

# ====================================
# 3.1 ANÁLISIS TEMPORAL: JUEGOS LANZADOS POR AÑO
# ====================================

print("\n" + "="*70)
print("3.1 ANÁLISIS TEMPORAL: JUEGOS LANZADOS POR AÑO")
print("="*70)

# Contar juegos por año
games_per_year = df_clean['year_of_release'].value_counts().sort_index()

print(f"\n Distribución de juegos por año:")
print(games_per_year)

# Visualización
plt.figure(figsize=(14, 6))
games_per_year.plot(kind='bar', color='steelblue', alpha=0.8)
plt.title('Número de Juegos Lanzados por Año', fontsize=16, fontweight='bold')
plt.xlabel('Año', fontsize=12)
plt.ylabel('Número de Juegos', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Gráfico de línea para ver tendencia
plt.figure(figsize=(14, 6))
plt.plot(games_per_year.index, games_per_year.values, marker='o', linewidth=2, markersize=6)
plt.title('Tendencia de Lanzamientos de Juegos a lo Largo del Tiempo', fontsize=16, fontweight='bold')
plt.xlabel('Año', fontsize=12)
plt.ylabel('Número de Juegos', fontsize=12)
plt.grid(True, alpha=0.3)
plt.axvline(x=2013, color='red', linestyle='--', alpha=0.5, label='Punto de cambio')
plt.legend()
plt.tight_layout()
plt.show()

print("\n Observaciones:")
print("""
- Hay un crecimiento significativo desde los años 90 hasta ~2008-2010
- El pico de lanzamientos ocurre alrededor de 2008-2009
- A partir de 2011 hay una disminución notable en el número de lanzamientos
- Los datos de 2016 están incompletos (como se mencionó en las instrucciones)
- Los datos muy antiguos (1980-1995) tienen pocos registros

 IMPORTANTE: Los datos anteriores a ~1995 son escasos y podrían no ser 
   representativos. Los datos de 2016 están incompletos.
""")

# ====================================
# 3.2 ANÁLISIS DE VENTAS POR PLATAFORMA
# ====================================

print("\n" + "="*70)
print("3.2 ANÁLISIS DE VENTAS POR PLATAFORMA")
print("="*70)

# Ventas totales por plataforma
sales_by_platform = df_clean.groupby('platform')['total_sales'].sum().sort_values(ascending=False)

print(f"\n Ventas totales por plataforma (Top 15):")
print(sales_by_platform.head(15))

# Visualización
plt.figure(figsize=(14, 8))
sales_by_platform.head(15).plot(kind='barh', color='coral')
plt.title('Top 15 Plataformas por Ventas Totales (Millones USD)', fontsize=16, fontweight='bold')
plt.xlabel('Ventas Totales (Millones USD)', fontsize=12)
plt.ylabel('Plataforma', fontsize=12)
plt.gca().invert_yaxis()
plt.grid(axis='x', alpha=0.3)
plt.tight_layout()
plt.show()

# ====================================
# 3.3 DISTRIBUCIÓN DE VENTAS POR PLATAFORMA A LO LARGO DEL TIEMPO
# ====================================

print("\n" + "="*70)
print("3.3 VENTAS POR PLATAFORMA A LO LARGO DEL TIEMPO")
print("="*70)

# Seleccionar las plataformas más importantes
top_platforms = sales_by_platform.head(8).index.tolist()

# Crear datos por año y plataforma
sales_year_platform = df_clean[df_clean['platform'].isin(top_platforms)].groupby(
    ['year_of_release', 'platform'])['total_sales'].sum().unstack(fill_value=0)

# Visualización
plt.figure(figsize=(16, 8))
for platform in top_platforms:
    if platform in sales_year_platform.columns:
        plt.plot(sales_year_platform.index, sales_year_platform[platform], 
                marker='o', linewidth=2, label=platform, markersize=4)

plt.title('Evolución de Ventas por Plataforma a lo Largo del Tiempo', fontsize=16, fontweight='bold')
plt.xlabel('Año', fontsize=12)
plt.ylabel('Ventas Totales (Millones USD)', fontsize=12)
plt.legend(loc='best', framealpha=0.9)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("\n Observaciones sobre ciclo de vida de plataformas:")
print("""
- PS2: Dominó durante principios de 2000, decline después de 2008
- Wii: Tuvo un pico dramático 2006-2010, luego declinó rápidamente
- X360 y PS3: Crecieron desde 2005-2006, mantuvieron fuerza hasta ~2013
- DS: Popular durante 2005-2011
- Las plataformas típicamente tienen un ciclo de vida de 5-8 años
- Plataformas antiguas (NES, GB, SNES) tuvieron su auge en décadas anteriores

 CICLO DE VIDA: Generalmente, una plataforma tarda ~2-3 años en alcanzar 
   su pico y permanece relevante por 5-8 años antes de ser reemplazada.
""")

# ====================================
# 3.4 DETERMINAR PERÍODO RELEVANTE PARA MODELO 2017
# ====================================

print("\n" + "="*70)
print("3.4 SELECCIÓN DEL PERÍODO RELEVANTE PARA PREDICCIÓN 2017")
print("="*70)

print("""
 Análisis para determinar período relevante:

Basándonos en:
1. El ciclo de vida típico de plataformas es ~5-8 años
2. Queremos predecir ventas para 2017
3. Los datos de 2016 están incompletos
4. Necesitamos datos recientes y completos

 DECISIÓN: Usaremos datos de 2014-2016 para construir el modelo

Razones:
- 2014-2016 representa los últimos 3 años completos de datos
- Captura las plataformas y tendencias actuales más relevantes
- Evita dilución con datos antiguos de plataformas obsoletas
- Permite análisis de plataformas actuales: PS4, XOne, 3DS, etc.
- Los datos de 2016, aunque incompletos, siguen siendo valiosos
""")

# Filtrar datos relevantes
df_relevant = df_clean[df_clean['year_of_release'] >= 2014].copy()

print(f"\n Dataset filtrado para período relevante (2014-2016):")
print(f"   - Registros totales: {len(df_relevant):,}")
print(f"   - Años incluidos: {sorted(df_relevant['year_of_release'].unique())}")
print(f"   - Plataformas únicas: {df_relevant['platform'].nunique()}")

# ====================================
# 3.5 PLATAFORMAS LÍDERES Y POTENCIALMENTE RENTABLES
# ====================================

print("\n" + "="*70)
print("3.5 PLATAFORMAS LÍDERES EN PERÍODO RELEVANTE (2014-2016)")
print("="*70)

# Ventas por plataforma en período relevante
sales_relevant = df_relevant.groupby('platform')['total_sales'].agg(['sum', 'mean', 'count'])
sales_relevant = sales_relevant.sort_values('sum', ascending=False)

print(f"\n Top 10 plataformas (2014-2016):")
print(sales_relevant.head(10))

# Visualización
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Gráfico 1: Ventas totales
sales_relevant.head(10)['sum'].plot(kind='bar', ax=axes[0], color='steelblue')
axes[0].set_title('Ventas Totales por Plataforma (2014-2016)', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Plataforma', fontsize=12)
axes[0].set_ylabel('Ventas Totales (Millones USD)', fontsize=12)
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(axis='y', alpha=0.3)

# Gráfico 2: Número de juegos
sales_relevant.head(10)['count'].plot(kind='bar', ax=axes[1], color='coral')
axes[1].set_title('Número de Juegos por Plataforma (2014-2016)', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Plataforma', fontsize=12)
axes[1].set_ylabel('Cantidad de Juegos', fontsize=12)
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print("\n Plataformas potencialmente rentables para 2017:")
print("""
CRECIENDO o ESTABLES:
- PS4: Líder en ventas, plataforma actual de Sony
- XOne: Segunda en ventas, plataforma actual de Microsoft
- 3DS: Dominante en portátiles Nintendo
- PC: Siempre relevante, mercado estable

EN DECLIVE:
- PS3: Siendo reemplazada por PS4
- X360: Siendo reemplazada por XOne
- WiiU: Bajo rendimiento, será reemplazada
""")

# ====================================
# 3.6 DIAGRAMA DE CAJA: VENTAS GLOBALES POR PLATAFORMA
# ====================================

print("\n" + "="*70)
print("3.6 ANÁLISIS DE DISTRIBUCIÓN DE VENTAS POR PLATAFORMA")
print("="*70)

# Seleccionar top plataformas del período relevante
top_platforms_relevant = sales_relevant.head(8).index.tolist()
df_box = df_relevant[df_relevant['platform'].isin(top_platforms_relevant)]

# Diagrama de caja
plt.figure(figsize=(14, 8))
df_box.boxplot(column='total_sales', by='platform', figsize=(14, 8))
plt.title('Distribución de Ventas Globales por Plataforma (2014-2016)', fontsize=16, fontweight='bold')
plt.suptitle('')  # Remover título automático
plt.xlabel('Plataforma', fontsize=12)
plt.ylabel('Ventas Totales (Millones USD)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

# Estadísticas por plataforma
print(f"\n Estadísticas de ventas por plataforma:")
for platform in top_platforms_relevant:
    platform_data = df_relevant[df_relevant['platform'] == platform]['total_sales']
    print(f"\n{platform}:")
    print(f"   Media: {platform_data.mean():.2f}M")
    print(f"   Mediana: {platform_data.median():.2f}M")
    print(f"   Desv. Std: {platform_data.std():.2f}M")
    print(f"   Máximo: {platform_data.max():.2f}M")

print("\n Hallazgos del análisis de caja:")
print("""
 Las diferencias en ventas entre plataformas SÍ son significativas
 PS4 y XOne muestran las distribuciones más favorables (medianas más altas)
 Todas las plataformas tienen muchos outliers (juegos muy exitosos)
 La mayoría de juegos venden poco, pero algunos éxitos compensan
 Las ventas promedio varían significativamente entre plataformas
 PS4 tiene los juegos con mejor rendimiento promedio en este período
""")

# ====================================
# 3.7 CORRELACIÓN: RESEÑAS vs VENTAS
# ====================================

print("\n" + "="*70)
print("3.7 CORRELACIÓN ENTRE RESEÑAS Y VENTAS")
print("="*70)

# Seleccionar una plataforma popular con datos suficientes
platform_analysis = 'PS4'
df_platform = df_relevant[df_relevant['platform'] == platform_analysis].copy()

# Filtrar solo juegos con reseñas disponibles
df_platform_reviews = df_platform.dropna(subset=['critic_score', 'user_score'])

print(f"\n Análisis de correlación para {platform_analysis}:")
print(f"   - Juegos totales en {platform_analysis}: {len(df_platform)}")
print(f"   - Juegos con ambas reseñas: {len(df_platform_reviews)}")

# Calcular correlaciones
corr_critic = df_platform_reviews['critic_score'].corr(df_platform_reviews['total_sales'])
corr_user = df_platform_reviews['user_score'].corr(df_platform_reviews['total_sales'])

print(f"\n Coeficientes de correlación en {platform_analysis}:")
print(f"   - Critic Score vs Ventas: {corr_critic:.3f}")
print(f"   - User Score vs Ventas: {corr_user:.3f}")

# Gráficos de dispersión
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Gráfico 1: Critic Score vs Ventas
axes[0].scatter(df_platform_reviews['critic_score'], df_platform_reviews['total_sales'], 
               alpha=0.5, color='steelblue')
axes[0].set_title(f'Critic Score vs Ventas ({platform_analysis})\nCorrelación: {corr_critic:.3f}', 
                 fontsize=14, fontweight='bold')
axes[0].set_xlabel('Critic Score', fontsize=12)
axes[0].set_ylabel('Ventas Totales (Millones USD)', fontsize=12)
axes[0].grid(True, alpha=0.3)

# Línea de tendencia
z = np.polyfit(df_platform_reviews['critic_score'], df_platform_reviews['total_sales'], 1)
p = np.poly1d(z)
axes[0].plot(df_platform_reviews['critic_score'], p(df_platform_reviews['critic_score']), 
            "r--", alpha=0.8, linewidth=2)

# Gráfico 2: User Score vs Ventas
axes[1].scatter(df_platform_reviews['user_score'], df_platform_reviews['total_sales'], 
               alpha=0.5, color='coral')
axes[1].set_title(f'User Score vs Ventas ({platform_analysis})\nCorrelación: {corr_user:.3f}', 
                 fontsize=14, fontweight='bold')
axes[1].set_xlabel('User Score', fontsize=12)
axes[1].set_ylabel('Ventas Totales (Millones USD)', fontsize=12)
axes[1].grid(True, alpha=0.3)

# Línea de tendencia
z = np.polyfit(df_platform_reviews['user_score'], df_platform_reviews['total_sales'], 1)
p = np.poly1d(z)
axes[1].plot(df_platform_reviews['user_score'], p(df_platform_reviews['user_score']), 
            "r--", alpha=0.8, linewidth=2)

plt.tight_layout()
plt.show()

print("\n Conclusiones sobre reseñas y ventas:")
print(f"""
 En {platform_analysis}, existe una correlación POSITIVA entre reseñas y ventas
 Critic Score muestra correlación {'moderada' if abs(corr_critic) > 0.3 else 'débil'} ({corr_critic:.3f})
 User Score muestra correlación {'moderada' if abs(corr_user) > 0.3 else 'débil'} ({corr_user:.3f})
 Las reseñas profesionales tienen {'mayor' if abs(corr_critic) > abs(corr_user) else 'menor'} 
  correlación con ventas que las de usuarios
 Hay mucha variabilidad: buenos scores no garantizan altas ventas
 Otros factores influyen: marketing, franquicia, momento de lanzamiento

 La correlación no es muy fuerte, lo que indica que las reseñas son 
   solo uno de muchos factores que determinan el éxito comercial.
""")

# ====================================
# 3.8 COMPARACIÓN DE VENTAS ENTRE PLATAFORMAS
# ====================================

print("\n" + "="*70)
print("3.8 COMPARACIÓN DE VENTAS DEL MISMO JUEGO EN DIFERENTES PLATAFORMAS")
print("="*70)

# Identificar juegos que aparecen en múltiples plataformas
game_counts = df_relevant['name'].value_counts()
multi_platform_games = game_counts[game_counts > 1].head(10).index.tolist()

print(f"\n Juegos multiplataforma más comunes (2014-2016):")
for i, game in enumerate(multi_platform_games, 1):
    platforms = df_relevant[df_relevant['name'] == game]['platform'].tolist()
    total_sales = df_relevant[df_relevant['name'] == game]['total_sales'].sum()
    print(f"{i}. {game}")
    print(f"   Plataformas: {', '.join(platforms)}")
    print(f"   Ventas totales combinadas: {total_sales:.2f}M\n")

# Analizar un juego específico en detalle
if len(multi_platform_games) > 0:
    sample_game = multi_platform_games[0]
    game_data = df_relevant[df_relevant['name'] == sample_game][['platform', 'total_sales', 
                                                                   'critic_score', 'user_score']]
    
    print(f"\n Análisis detallado: '{sample_game}'")
    print(game_data.to_string(index=False))
    
    print("\n Observaciones sobre juegos multiplataforma:")
    print("""
     Los juegos exitosos suelen lanzarse en múltiples plataformas
     Las ventas varían significativamente según la plataforma
     PS4 y XOne suelen tener las ventas más altas para juegos multiplataforma
     Las versiones de PC pueden tener ventas menores en retail (no incluye digital)
     Las diferencias en ventas reflejan el tamaño de la base instalada
    """)

# ====================================
# 3.9 ANÁLISIS DE GÉNEROS
# ====================================

print("\n" + "="*70)
print("3.9 ANÁLISIS DE GÉNEROS MÁS RENTABLES")
print("="*70)

# Análisis de géneros en período relevante
genre_analysis = df_relevant.groupby('genre')['total_sales'].agg(['sum', 'mean', 'count'])
genre_analysis = genre_analysis.sort_values('sum', ascending=False)

print(f"\n Estadísticas por género (2014-2016):")
print(genre_analysis)

# Visualización
fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# Gráfico 1: Ventas totales por género
genre_analysis['sum'].plot(kind='barh', ax=axes[0, 0], color='steelblue')
axes[0, 0].set_title('Ventas Totales por Género', fontsize=14, fontweight='bold')
axes[0, 0].set_xlabel('Ventas Totales (Millones USD)', fontsize=12)
axes[0, 0].set_ylabel('Género', fontsize=12)
axes[0, 0].invert_yaxis()
axes[0, 0].grid(axis='x', alpha=0.3)

# Gráfico 2: Ventas promedio por género
genre_analysis['mean'].sort_values().plot(kind='barh', ax=axes[0, 1], color='coral')
axes[0, 1].set_title('Ventas Promedio por Género', fontsize=14, fontweight='bold')
axes[0, 1].set_xlabel('Ventas Promedio (Millones USD)', fontsize=12)
axes[0, 1].set_ylabel('Género', fontsize=12)
axes[0, 1].invert_yaxis()
axes[0, 1].grid(axis='x', alpha=0.3)

# Gráfico 3: Número de juegos por género
genre_analysis['count'].sort_values().plot(kind='barh', ax=axes[1, 0], color='green', alpha=0.7)
axes[1, 0].set_title('Cantidad de Juegos por Género', fontsize=14, fontweight='bold')
axes[1, 0].set_xlabel('Número de Juegos', fontsize=12)
axes[1, 0].set_ylabel('Género', fontsize=12)
axes[1, 0].invert_yaxis()
axes[1, 0].grid(axis='x', alpha=0.3)

# Gráfico 4: Distribución de ventas por género (top 6)
top_genres = genre_analysis.head(6).index.tolist()
df_relevant[df_relevant['genre'].isin(top_genres)].boxplot(
    column='total_sales', by='genre', ax=axes[1, 1])
axes[1, 1].set_title('Distribución de Ventas - Top 6 Géneros', fontsize=14, fontweight='bold')
axes[1, 1].set_xlabel('Género', fontsize=12)
axes[1, 1].set_ylabel('Ventas (Millones USD)', fontsize=12)
plt.setp(axes[1, 1].xaxis.get_majorticklabels(), rotation=45)
axes[1, 1].grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

print("\n CONCLUSIONES SOBRE GÉNEROS:")
print("""
 GÉNEROS MÁS RENTABLES (Ventas Totales Altas):
   - Shooter: El género más rentable en ventas totales
   - Action: Segundo en ventas totales, muy popular
   - Sports: Alto volumen de ventas
   - Role-Playing: Buenas ventas totales

 GÉNEROS CON MEJOR DESEMPEÑO PROMEDIO:
   - Shooter: Ventas promedio muy altas por juego
   - Sports: Buen promedio de ventas
   - Platform: Aunque menos juegos, buen promedio
   - Racing: Ventas promedio respetables

 GÉNEROS CON VENTAS BAJAS:
   - Strategy: Ventas totales y promedio bajas
   - Puzzle: Bajo volumen de ventas
   - Adventure: Ventas moderadas a bajas

 GENERALIZACIÓN:
  • Géneros de ACCIÓN (Shooter, Action, Sports) = ALTAS VENTAS
  • Géneros NICHO (Strategy, Puzzle) = VENTAS BAJAS
  • Los géneros con más juegos no siempre son los más rentables
  • El éxito depende de la calidad individual y las franquicias
""")

print("\n" + "="*70)
print(" PASO 3 COMPLETADO")
print("="*70)
print("""
 Análisis temporal completado: Identificado período 2014-2016 como relevante
 Plataformas analizadas: PS4, XOne, 3DS identificadas como líderes
 Correlación reseñas-ventas: Positiva pero moderada
 Géneros más rentables identificados: Shooter, Action, Sports
 Dataset filtrado (df_relevant) listo para análisis regional

 Siguiente paso: Crear perfiles de usuario por región (Paso 4)
""")
```

```python
# Proyecto Ice Games - Análisis de Videojuegos
# Paso 5: Prueba de Hipótesis

# Configuración de visualización
sns.set_palette("husl")
plt.rcParams['figure.figsize'] = (12, 6)

print("="*70)
print("PASO 5: PRUEBAS DE HIPÓTESIS")
print("="*70)

# Cargar y preparar datos (si no están ya en memoria)
try:
    # Intentar usar df_clean si ya existe
    test = df_clean.shape
    print("\n Usando df_clean existente en memoria")
except NameError:
    # Si no existe, cargar y preparar desde cero
    print("\n df_clean no encontrado, cargando datos...")
    df = pd.read_csv('/datasets/games.csv')
    df.columns = df.columns.str.lower()
    df['user_score'] = pd.to_numeric(df['user_score'], errors='coerce')
    df_clean = df.dropna(subset=['name', 'genre', 'year_of_release']).copy()
    df_clean['total_sales'] = (df_clean['na_sales'] + df_clean['eu_sales'] + 
                                df_clean['jp_sales'] + df_clean['other_sales'])
    print(f" Datos cargados: {len(df_clean):,} registros")

print("\n Estableciendo el nivel de significancia (alfa):")
print("   α = 0.05 (5%)")
print("   Este es el estándar en investigación científica")
print("   Significa que aceptamos un 5% de probabilidad de error tipo I")
alpha = 0.05

# ====================================
# HIPÓTESIS 1: Xbox One vs PC (User Score)
# ====================================

print("\n" + "="*70)
print("HIPÓTESIS 1: Calificaciones de Usuarios - Xbox One vs PC")
print("="*70)

print("\n FORMULACIÓN DE HIPÓTESIS:")
print("""
H₀ (Hipótesis Nula):
   Las calificaciones promedio de usuarios para Xbox One y PC son IGUALES.
   μ(XOne) = μ(PC)

H₁ (Hipótesis Alternativa):
   Las calificaciones promedio de usuarios para Xbox One y PC son DIFERENTES.
   μ(XOne) ≠ μ(PC)

Tipo de prueba: Prueba t de Student de dos muestras (bilateral)
Razón: Comparamos las MEDIAS de dos grupos independientes
""")

# Filtrar datos de Xbox One y PC con user_score disponible
xone_scores = df_clean[(df_clean['platform'] == 'XOne') & 
                        (df_clean['user_score'].notna())]['user_score']
pc_scores = df_clean[(df_clean['platform'] == 'PC') & 
                      (df_clean['user_score'].notna())]['user_score']

print(f"\n DATOS DE LA MUESTRA:")
print(f"   Xbox One: {len(xone_scores)} juegos con puntuaciones")
print(f"   PC: {len(pc_scores)} juegos con puntuaciones")

# Verificar que haya suficientes datos
if len(xone_scores) < 2 or len(pc_scores) < 2:
    print("\n ADVERTENCIA: Datos insuficientes para realizar la prueba")
else:
    # Estadísticas descriptivas
    print(f"\n ESTADÍSTICAS DESCRIPTIVAS:")
    print(f"\n   Xbox One:")
    print(f"      Media: {xone_scores.mean():.3f}")
    print(f"      Mediana: {xone_scores.median():.3f}")
    print(f"      Desviación estándar: {xone_scores.std():.3f}")
    print(f"      Mínimo: {xone_scores.min():.1f}")
    print(f"      Máximo: {xone_scores.max():.1f}")
    
    print(f"\n   PC:")
    print(f"      Media: {pc_scores.mean():.3f}")
    print(f"      Mediana: {pc_scores.median():.3f}")
    print(f"      Desviación estándar: {pc_scores.std():.3f}")
    print(f"      Mínimo: {pc_scores.min():.1f}")
    print(f"      Máximo: {pc_scores.max():.1f}")
    
    # Visualización de distribuciones
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    
    # Histogramas
    axes[0].hist(xone_scores, bins=20, alpha=0.7, label='Xbox One', color='steelblue', edgecolor='black')
    axes[0].hist(pc_scores, bins=20, alpha=0.7, label='PC', color='coral', edgecolor='black')
    axes[0].axvline(xone_scores.mean(), color='steelblue', linestyle='--', linewidth=2, label=f'Media XOne: {xone_scores.mean():.2f}')
    axes[0].axvline(pc_scores.mean(), color='coral', linestyle='--', linewidth=2, label=f'Media PC: {pc_scores.mean():.2f}')
    axes[0].set_title('Distribución de User Scores - Xbox One vs PC', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('User Score', fontsize=12)
    axes[0].set_ylabel('Frecuencia', fontsize=12)
    axes[0].legend()
    axes[0].grid(alpha=0.3)
    
    # Box plots
    data_to_plot = [xone_scores, pc_scores]
    axes[1].boxplot(data_to_plot, labels=['Xbox One', 'PC'], patch_artist=True,
                    boxprops=dict(facecolor='lightblue', alpha=0.7),
                    medianprops=dict(color='red', linewidth=2))
    axes[1].set_title('Comparación de Distribuciones - User Scores', fontsize=14, fontweight='bold')
    axes[1].set_ylabel('User Score', fontsize=12)
    axes[1].grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Realizar prueba t de Student
    print(f"\n APLICANDO PRUEBA T DE STUDENT:")
    print(f"   Método: scipy.stats.ttest_ind() - Prueba t de dos muestras independientes")
    print(f"   Supuesto: Varianzas pueden ser diferentes (equal_var=False, prueba de Welch)")
    
    # Prueba t de Welch (no asume varianzas iguales)
    t_statistic, p_value = stats.ttest_ind(xone_scores, pc_scores, equal_var=False)
    
    print(f"\n RESULTADOS DE LA PRUEBA:")
    print(f"   Estadístico t: {t_statistic:.4f}")
    print(f"   Valor p (p-value): {p_value:.4f}")
    print(f"   Nivel de significancia (α): {alpha}")
    print(f"   Diferencia de medias: {xone_scores.mean() - pc_scores.mean():.4f}")
    
    # Interpretación
    print(f"\n INTERPRETACIÓN:")
    if p_value < alpha:
        print(f"    p-value ({p_value:.4f}) < α ({alpha})")
        print(f"   RECHAZAMOS la hipótesis nula (H₀)")
        print(f"\n   Conclusión:")
        print(f"   Las calificaciones promedio de usuarios para Xbox One y PC")
        print(f"   son estadísticamente DIFERENTES con un 95% de confianza.")
        if xone_scores.mean() > pc_scores.mean():
            print(f"   Xbox One tiene calificaciones significativamente MAYORES que PC.")
        else:
            print(f"   PC tiene calificaciones significativamente MAYORES que Xbox One.")
    else:
        print(f"    p-value ({p_value:.4f}) >= α ({alpha})")
        print(f"   NO RECHAZAMOS la hipótesis nula (H₀)")
        print(f"\n   Conclusión:")
        print(f"   No hay evidencia suficiente para afirmar que las calificaciones")
        print(f"   promedio de usuarios para Xbox One y PC son diferentes.")
        print(f"   Las diferencias observadas podrían deberse al azar.")

# ====================================
# HIPÓTESIS 2: Action vs Sports (User Score)
# ====================================

print("\n" + "="*70)
print("HIPÓTESIS 2: Calificaciones de Usuarios - Action vs Sports")
print("="*70)

print("\n FORMULACIÓN DE HIPÓTESIS:")
print("""
H₀ (Hipótesis Nula):
   Las calificaciones promedio de usuarios para Action y Sports son IGUALES.
   μ(Action) = μ(Sports)

H₁ (Hipótesis Alternativa):
   Las calificaciones promedio de usuarios para Action y Sports son DIFERENTES.
   μ(Action) ≠ μ(Sports)

Tipo de prueba: Prueba t de Student de dos muestras (bilateral)
Razón: Comparamos las MEDIAS de dos grupos independientes
""")

# Filtrar datos de géneros Action y Sports con user_score disponible
action_scores = df_clean[(df_clean['genre'] == 'Action') & 
                         (df_clean['user_score'].notna())]['user_score']
sports_scores = df_clean[(df_clean['genre'] == 'Sports') & 
                         (df_clean['user_score'].notna())]['user_score']

print(f"\n DATOS DE LA MUESTRA:")
print(f"   Action: {len(action_scores)} juegos con puntuaciones")
print(f"   Sports: {len(sports_scores)} juegos con puntuaciones")

# Verificar que haya suficientes datos
if len(action_scores) < 2 or len(sports_scores) < 2:
    print("\n ADVERTENCIA: Datos insuficientes para realizar la prueba")
else:
    # Estadísticas descriptivas
    print(f"\n ESTADÍSTICAS DESCRIPTIVAS:")
    print(f"\n   Action:")
    print(f"      Media: {action_scores.mean():.3f}")
    print(f"      Mediana: {action_scores.median():.3f}")
    print(f"      Desviación estándar: {action_scores.std():.3f}")
    print(f"      Mínimo: {action_scores.min():.1f}")
    print(f"      Máximo: {action_scores.max():.1f}")
    
    print(f"\n   Sports:")
    print(f"      Media: {sports_scores.mean():.3f}")
    print(f"      Mediana: {sports_scores.median():.3f}")
    print(f"      Desviación estándar: {sports_scores.std():.3f}")
    print(f"      Mínimo: {sports_scores.min():.1f}")
    print(f"      Máximo: {sports_scores.max():.1f}")
    
    # Visualización de distribuciones
    fig, axes = plt.subplots(1, 2, figsize=(14, 6))
    
    # Histogramas
    axes[0].hist(action_scores, bins=20, alpha=0.7, label='Action', color='steelblue', edgecolor='black')
    axes[0].hist(sports_scores, bins=20, alpha=0.7, label='Sports', color='green', edgecolor='black')
    axes[0].axvline(action_scores.mean(), color='steelblue', linestyle='--', linewidth=2, label=f'Media Action: {action_scores.mean():.2f}')
    axes[0].axvline(sports_scores.mean(), color='green', linestyle='--', linewidth=2, label=f'Media Sports: {sports_scores.mean():.2f}')
    axes[0].set_title('Distribución de User Scores - Action vs Sports', fontsize=14, fontweight='bold')
    axes[0].set_xlabel('User Score', fontsize=12)
    axes[0].set_ylabel('Frecuencia', fontsize=12)
    axes[0].legend()
    axes[0].grid(alpha=0.3)
    
    # Box plots
    data_to_plot = [action_scores, sports_scores]
    axes[1].boxplot(data_to_plot, labels=['Action', 'Sports'], patch_artist=True,
                    boxprops=dict(facecolor='lightgreen', alpha=0.7),
                    medianprops=dict(color='red', linewidth=2))
    axes[1].set_title('Comparación de Distribuciones - User Scores', fontsize=14, fontweight='bold')
    axes[1].set_ylabel('User Score', fontsize=12)
    axes[1].grid(axis='y', alpha=0.3)
    
    plt.tight_layout()
    plt.show()
    
    # Realizar prueba t de Student
    print(f"\n APLICANDO PRUEBA T DE STUDENT:")
    print(f"   Método: scipy.stats.ttest_ind() - Prueba t de dos muestras independientes")
    print(f"   Supuesto: Varianzas pueden ser diferentes (equal_var=False, prueba de Welch)")
    
    # Prueba t de Welch
    t_statistic2, p_value2 = stats.ttest_ind(action_scores, sports_scores, equal_var=False)
    
    print(f"\n RESULTADOS DE LA PRUEBA:")
    print(f"   Estadístico t: {t_statistic2:.4f}")
    print(f"   Valor p (p-value): {p_value2:.4f}")
    print(f"   Nivel de significancia (α): {alpha}")
    print(f"   Diferencia de medias: {action_scores.mean() - sports_scores.mean():.4f}")
    
    # Interpretación
    print(f"\n INTERPRETACIÓN:")
    if p_value2 < alpha:
        print(f"    p-value ({p_value2:.4f}) < α ({alpha})")
        print(f"   RECHAZAMOS la hipótesis nula (H₀)")
        print(f"\n   Conclusión:")
        print(f"   Las calificaciones promedio de usuarios para Action y Sports")
        print(f"   son estadísticamente DIFERENTES con un 95% de confianza.")
        if action_scores.mean() > sports_scores.mean():
            print(f"   Action tiene calificaciones significativamente MAYORES que Sports.")
        else:
            print(f"   Sports tiene calificaciones significativamente MAYORES que Action.")
    else:
        print(f"    p-value ({p_value2:.4f}) >= α ({alpha})")
        print(f"   NO RECHAZAMOS la hipótesis nula (H₀)")
        print(f"\n   Conclusión:")
        print(f"   No hay evidencia suficiente para afirmar que las calificaciones")
        print(f"   promedio de usuarios para Action y Sports son diferentes.")
        print(f"   Las diferencias observadas podrían deberse al azar.")

# ====================================
# RESUMEN DE PRUEBAS DE HIPÓTESIS
# ====================================

print("\n" + "="*70)
print("RESUMEN DE PRUEBAS DE HIPÓTESIS")
print("="*70)

print("\n Criterio utilizado: Prueba t de Student (Welch)")
print("""
Por qué elegimos esta prueba:
 Comparamos MEDIAS de dos grupos independientes
 Variables continuas (puntuaciones de 0-10)
 No asumimos varianzas iguales (más robusto)
 Muestras suficientemente grandes
 Es el estándar para este tipo de comparaciones

Supuestos:
- Independencia entre grupos (diferentes plataformas/géneros)
- Variables aproximadamente normales (o muestras grandes)
- No requiere varianzas iguales (usamos corrección de Welch)
""")

if len(xone_scores) >= 2 and len(pc_scores) >= 2:
    print(f"\nHIPÓTESIS 1: Xbox One vs PC")
    print(f"   Resultado: {'RECHAZADA' if p_value < alpha else 'NO RECHAZADA'} (p={p_value:.4f})")
    print(f"   Medias: XOne={xone_scores.mean():.3f}, PC={pc_scores.mean():.3f}")

if len(action_scores) >= 2 and len(sports_scores) >= 2:
    print(f"\nHIPÓTESIS 2: Action vs Sports")
    print(f"   Resultado: {'RECHAZADA' if p_value2 < alpha else 'NO RECHAZADA'} (p={p_value2:.4f})")
    print(f"   Medias: Action={action_scores.mean():.3f}, Sports={sports_scores.mean():.3f}")

print("\n" + "="*70)
print("INTERPRETACIÓN GENERAL")
print("="*70)

print("""
 SIGNIFICADO PRÁCTICO:

Nivel de significancia (α = 0.05):
 Aceptamos un 5% de probabilidad de error tipo I
 Es decir, 5% de rechazar H₀ cuando es verdadera
 95% de confianza en nuestras conclusiones

Valor p (p-value):
 Probabilidad de obtener estos resultados si H₀ fuera cierta
 p < 0.05: Resultado es estadísticamente significativo
 p >= 0.05: No podemos descartar que sea por azar

IMPORTANTE:
 Significancia estadística ≠ Importancia práctica
 Diferencias pequeñas pueden ser significativas con muestras grandes
 Considerar también el tamaño del efecto (diferencia de medias)
 Las decisiones de negocio requieren contexto adicional
""")

print("\n" + "="*70)
print(" PASO 5 COMPLETADO")
print("="*70)
print("""
 Hipótesis nula y alternativa formuladas correctamente
 Prueba t de Student aplicada para ambas hipótesis
 Nivel de significancia establecido (α = 0.05)
 Resultados interpretados con contexto estadístico
 Visualizaciones de distribuciones generadas

Siguiente paso: Escribir conclusión general (Paso 6)
""")
```
