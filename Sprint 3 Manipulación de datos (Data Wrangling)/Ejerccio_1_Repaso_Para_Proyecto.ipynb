### 3.1 Uso de `.loc[]` en Pandas

El m칠todo `.loc[]` permite **seleccionar y asignar** datos en un DataFrame utilizando **etiquetas** (칤ndices y nombres de columna), en lugar de posiciones num칠ricas.

#### Sintaxis general
```python
df.loc[filtro_de_filas, selecci칩n_de_columnas]
```
Par치metros
- filtro_de_filas
    - Etiqueta 칰nica: una sola etiqueta de 칤ndice (df.loc[5]).
    - Lista de etiquetas: varias etiquetas (df.loc[[1, 3, 5]]).
    - Condici칩n booleana: mascarilla que devuelve True/False por fila (df.loc[df["edad"] > 30]).
selecci칩n_de_columnas (opcional)
    - Etiqueta 칰nica: un nombre de columna (df.loc[:, "nombre"]).
    - Lista de etiquetas: varias columnas (df.loc[:, ["nombre", "edad"]]).
    - Rango de etiquetas: mediante slicing (df.loc[:, "col1":"col4"]).
    - Si se omite, se seleccionan todas las columnas (df.loc[filtro_de_filas]).

Ejemplos de uso

```python
# 1. Filtrar por condici칩n y ver todas las columnas:
df.loc[df["ventas"] > 1000]

# 2. Filtrar filas y ver solo ciertas columnas:
df.loc[df["categoria"] == "Electr칩nica", ["producto", "precio"]]

# 3. Selecci칩n por etiquetas de fila y columna:
df.loc["fila_10", "columna_A"]

# 4. Asignaci칩n de valores a un subconjunto:
df.loc[df["puntuacion"] < 50, "estado"] = "revisar"
```

**Tip:** .loc respeta los nombres de 칤ndices y columnas. Para selecci칩n basada en posiciones, utiliza .iloc[].

````python
# Seleccionar la fila con 칤ndice 5 (sexta fila)
df.loc[5]
````

### 3.2 Uso de `.iloc[]` en Pandas

```python
df.iloc[fila_inicio:fila_fin, columna_inicio:columna_fin]
```
- Qu칠 hace: Selecciona filas y/o columnas por posiciones enteras (integer-location), ignorando las etiquetas del 칤ndice.
- Sintaxis b치sica:
    - df.iloc[5]  quinta fila (0-based) en todas las columnas.
    - df.iloc[2:8]  filas de la 3 a la 8 (excluye la 8).
    - df.iloc[:, 1]  segunda columna en todas las filas.
    - df.iloc[3:6, 0:2]  filas 4-6 y columnas 1-2.

**Uso t칤pico:** Cuando se necesita seleccionar por posiciones num칠ricas o hacer slicing posicional en lugar de por etiquetas.

````python
df.iloc[:, 1:4]
````

````python
df.iloc[6:, 7:]
````

### 3.4 Operaciones de filtrado en Pandas

```python
df[df['Album'] == '1989']
```
- Selecciona todas las filas donde la columna **Album** es igual a `'1989'`.

```python
df[(df['Spotify Streams (millions)'] >= 900) & (df['Spotify Streams (millions)'] <= 1000)]
```
- Selecciona filas donde **Spotify Streams (millions)** est치 entre 900 y 1000 (inclusive).  
- Utiliza el operador **&** para combinar condiciones booleanas y par칠ntesis para agrupar cada condici칩n.

> Estas t칠cnicas se conocen como **indexaci칩n booleana** o **filtrado condicional**, y devuelven un DataFrame con solo las filas que cumplen los criterios especificados.


````python
# Filtrar canciones que pertenecen al 치lbum '1989'
df_filtrado = df[df['Album'] == '1989']
df_filtrado
````

````python
# Filtrar canciones con reproducciones entre 500 y 1000 millones
# and = &
# or = |
df[(df['Genre'] == "Pop") &
   (df['Peak Billboard Position'] == 1) &
   (df['Duration (mins)'] >= 3)]
````

### 3.7 Operaciones de agrupamiento con `groupby` en Pandas

- **Qu칠 hace**: Agrupa filas del DataFrame seg칰n valores de una o varias columnas, devolviendo un objeto `GroupBy` sobre el que se pueden aplicar funciones de agregaci칩n, transformaci칩n o filtrado.

- **Sintaxis b치sica**:
```python
df.groupby('columna')
df.groupby(['col1', 'col2'])
```

- **M칠todos comunes**:
  - `agg()` o `aggregate()`: aplica funciones agregadas.
  - `mean()`, `sum()`, `count()`, `min()`, `max()`.
  - `size()`: devuelve el tama침o de cada grupo.
  - `get_group(nombre_grupo)`: obtiene el subconjunto correspondiente a un grupo espec칤fico.

- **Ejemplos**:
```python
# 1. Suma de ventas por categor칤a
df.groupby('categoria')['ventas'].sum()

# 2. Estad칤sticas descriptivas por regi칩n
df.groupby('region').describe()

# 3. N칰mero de registros por usuario
df.groupby('usuario').size()

# 4. Aplicar m칰ltiples agregaciones a ventas
df.groupby('categoria')['ventas'].agg(['sum', 'mean', 'max'])
```

````python
df.groupby('Release Year')['Title'].count()
````

## 4. Manejo de valores nulos

Es esencial que cualquier cient칤fico de datos pueda manejar valores nulos.  
Los valores nulos no nos aportan informaci칩n y, en ocasiones, pueden causar errores en el an치lisis o en el entrenamiento de modelos.  

En **pandas**, los valores nulos suelen representarse como:
- `NaN` (Not a Number)  
- `None` (en objetos tipo string)  
- `NaT` (para fechas)


````python
# Saber si hay valores nulos
df.isnull()
````

````python
# Contar valores nulos por columna
df.isnull().sum()
````

````python
# Comprobar si existe al menos un valor nulo
df.isnull().values.any()
````

````python
# Eliminar filas con valores nulos
df.dropna()
````

````python
# Eliminar columnas con valores nulos
df.dropna(axis=1)
````

````python
# Rellenar con un valor fijo
df.fillna(0)

# Rellenar con el valor anterior (forward fill)
df.fillna(method="ffill")

# Rellenar con el siguiente valor (backward fill)
df.fillna(method="bfill")

# Rellenar con la media de la columna
# df["edad"].fillna(df["edad"].mean(), inplace=True) no aplica en nuestro caso
````
